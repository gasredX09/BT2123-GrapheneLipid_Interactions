```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Set the figure size
plt.rcParams["figure.figsize"] = [7.50, 3.50]
plt.rcParams["figure.autolayout"] = True
```

```python
#Loading the data from the .dat file
data = np.loadtxt("lipid_in_contact.dat")
```

```python
#Extracting the x and y data
x = data[:, 0]
y = data[:, 1]

plt.title("Lipids in Contact vs Time")
plt.xlabel("Time(in ns)")
plt.ylabel("Number of Lipids in Contact with Graphene")

from scipy import interpolate
x_y_spline = interpolate.make_interp_spline(x, y)

x_ = np.linspace(x.min(), x.max())
y_ = x_y_spline(x_)

#Creating a line plot
plt.plot(x, y, color = 'orange')

#Creating a smooth plot
plt.plot(x_, y_, color = 'purple')
plt.show()
```
